# Calculating Ranges

```python
currencies = set([col.split('-')[0] for col in df_x.columns if '-high' in col or '-low' in col])

for currency in currencies:
    high_col = f"{currency}-high"
    low_col = f"{currency}-low"
    diff_col = f"{currency}-volatility"
    df_x[diff_col] = df_x[high_col] - df_x[low_col]
```

```python
all_columns = df_x.columns

# Extract columns with "-high" suffix
vol_columns = [col for col in all_columns if '-volatility' in col]

# Include the 'date' column
selected_columns = ['date'] + vol_columns

# Create the subdataframe
sub_df = df_x[selected_columns]
```

```python
correlations_vol = sub_df.corrwith(sub_df['CAD-volatility']).drop('CAD-volatility')
```
